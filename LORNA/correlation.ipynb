```python
import pandas as pd
import geopandas as gpd
shape=gpd.read_file('Outputs/County_Data.shp')
```

```python
data=pd.DataFrame()
data[['Victims','Income','ICT_Access','Male','Female','Intersex','Total','Electricity']]=shape[['victims','Income_Gin','ICT_Access','Male','Female','Intersex','Total','Electricit']]
```

```python
inc=[]
for i in data['Income']:
    a=i.split("%")
    b=a[0]
    inc.append(b)
incint=[]
for j in inc:
    a=int(float(j))
    incint.append(a)
```

```python
data['income']=incint
```
